```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
df['Age'].describe()
```

```python
sns.histplot(df['Age'], bins=20)
plt.title("Age Distribution of Employees")
plt.show()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Attrition'].value_counts()
```

```python
sns.countplot(x='Attrition', data=df)
plt.title("Employee Attrition Count")
plt.show()
```

```python
df['Age'].describe()
```

```python
sns.histplot(data=df, x='Age', hue='Attrition', multiple='stack')
plt.title("Age Distribution by Attrition")
plt.show()
```

```python
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Salary vs Attrition")
plt.show()
```

```python
dept_attrition = pd.crosstab(df['Department'], df['Attrition'])
dept_attrition.plot(kind='bar', figsize=(8,5))
plt.title("Department Wise Attrition")
plt.show()
```

```python
job_attrition = pd.crosstab(df['JobRole'], df['Attrition'])
job_attrition.plot(kind='bar', figsize=(12,6))
plt.title("Job Role vs Attrition")
plt.show()
```

```python
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title("Overtime vs Attrition")
plt.show()
```

```python
print("Key Insights:")
print("- Younger employees show higher attrition")
print("- Lower salary employees leave more")
print("- Sales department attrition is high")
print("- Overtime strongly impacts attrition")
```
